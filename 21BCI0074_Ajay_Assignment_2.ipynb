```python
import seaborn as sns
import matplotlib.pyplot as plt

car_crashes = sns.load_dataset("car_crashes")
car_crashes.head()
```

```python
car_crashes.tail()
```

```python
# Correlation Heatmap
corr_matrix = car_crashes.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

**Inference for Correlation Heatmap:**

The above graph is Heatmap. it shows the correlation between numerical variables. For example, "alcohol" and "total" have a strong positive correlation.


```python
#Box Plot
sns.boxplot(x="total", data=car_crashes, color="gold")
plt.title("Box Plot of Total Crashes")
plt.xlabel("Total Crashes")
plt.show()
```

**Inference for Box Plot:**


The box plot provides a graphical representation of the total crash data distribution while also helping to pinpoint potential anomalies or outliers. The central tendency, represented by the median, falls at approximately 17 total crashes.

```python
# Line Plot
sns.lineplot(x="total", y="alcohol", data=car_crashes, marker="o", color="gold")
plt.title("Line Plot of Total Crashes vs. Alcohol-Related Crashes")
plt.xlabel("Total Crashes")
plt.ylabel("Alcohol-Related Crashes")
plt.show()
```

**Inference for Line Plot:**

The line chart depicts a correlation between the total number of accidents and the quantity of alcohol-related accidents. As the overall number of accidents rises, there is a corresponding increase in the occurrences of alcohol-related accidents. This indicates a favorable connection between these two variables.


```python
#Scatter Plot
sns.scatterplot(x="alcohol", y="speeding", data=car_crashes, color = "gold")
plt.title("Scatter Plot of Alcohol vs. Speeding")
plt.xlabel("Alcohol-related Crashes")
plt.ylabel("Speeding-related Crashes")
plt.show()
```

**Inference for Scatter Plot:**

The scatter plot illustrates a positive correlation between crashes related to alcohol consumption and those associated with speeding.


```python
#Bar Plot
sns.barplot(x="total", y="abbrev", data=car_crashes.sort_values(by="total", ascending=False))
plt.title("Total Crashes by State")
plt.xlabel("Total Crashes")
plt.ylabel("State (Abbreviation)")
plt.yticks(fontsize=5)
plt.show()
```

**Inference for Bar Plot:**

This bar chart illustrates the total crash counts per state, with SC having the highest number of accidents and DC recording the lowest total crash incidents.


```python
# Histogram
plt.hist(car_crashes["total"], bins=10, color="gold")
plt.title("Histogram of Total Crashes")
plt.xlabel("Total Crashes")
plt.ylabel("Frequency")
plt.show()
```

**Inference for Histogram:**

The above graph is Histogram. This histogram displays the distribution of total crashes. It appears to be right-skewed, with most states having a lower number of total crashes.
